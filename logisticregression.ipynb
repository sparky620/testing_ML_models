```python
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('HR_comma_sep.csv')
df
```

```python
plt.scatter( df['satisfaction_level'],df['left'],marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(df[['satisfaction_level','average_montly_hours','number_project','last_evaluation','time_spend_company']],df.left,test_size=0.2)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```
